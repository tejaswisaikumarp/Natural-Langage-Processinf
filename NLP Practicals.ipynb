```python
import re
```

```python
details_one = "This is my mobile_numbers and email_id: (012)-345-6789, 7418529630, abc@yahoo.com, ab_1_CD@gmail.com"
```

```python
'''
Source: https://regex101.com/

\d is a single number
\d{x} is a group of characters. x = 1,2,3.....

[a-z] is a single character
[a-z]* is a group of characters

 '''
pattern_mobile_number = ' \d{10}|\(\d{3}\)-\d{3}-\d{4}'
pattern_email_id = '[a-zA-Z0-9_]*@[a-zA-Z0-9_]*\.[a-z]*'
match_mobile_number = re.findall(pattern_mobile_number, details_one)
match_email_id = re.findall(pattern_email_id, details_one)
print(match_mobile_number)
print(match_email_id)
```

```python
'''
[^\d] - except digit
Question: order_number is 567. I want only 567 to be printed as an output. How ?
Answer: order_number[^\d]*([\d]*) - You need to place () so that only number gets printed.

'''
chat_one = "My product number is 123456789 and the other product # is 987654321"

pattern_product_number = "product[^\d]*([\d]*)"

match_product_number = re.findall(pattern_product_number, chat_one)
print(match_product_number)
```

```python
'''Information Extraction'''

pattern_age = 'age ([\d]+)'
pattern_born = 'Born(.*)\n'
pattern_dob = 'Born.*\n(.*)\(age'
text ='''
Born	             TejaswiSaiKumar
                   15 July 1997 (age 26)
                   Guntur, Andhra Pradesh, India
Other names        Teja
Occupation	       Student
Years active       1997-till date
'''

match_born = re.findall(pattern_born, text)
match_age = re.findall(pattern_age, text)
match_dob = re.findall(pattern_dob, text)

print(match_born)
'Add strip() to avoid spaces'
print(match_born[0].strip(), "\n")

print(match_age, "\n")
print(match_dob)
print(match_dob[0].strip())
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def preprocess(text):
  doc = nlp(text)
  token_filter = []
  for token in doc:
    if token.is_stop or token.is_punct:
      continue
    token_filter.append(token.lemma_)

  return " ".join(token_filter)
```

```python
import pandas as pd
df = pd.read_json('/content/news_dataset.json')
print(df.shape, "\n")
df.head(10)
```

```python
minimum_samples = 1381
df_business = df[df.category == "BUSINESS"].sample(minimum_samples, random_state = 42)
df_sports = df[df.category == "SPORTS"].sample(minimum_samples, random_state = 42)
df_crime = df[df.category == "CRIME"].sample(minimum_samples, random_state = 42)
df_science = df[df.category == "SCIENCE"].sample(minimum_samples, random_state = 42)
```

```python
balanced_data = pd.concat([df_business, df_sports, df_crime, df_science], axis = 0)
balanced_data.category.value_counts()
```

```python
target = {'BUSINESS': 0, 'SPORTS': 1, 'CRIME': 2, 'SCIENCE': 3}

balanced_data['category_number'] = balanced_data.category.map(target)
```

```python
balanced_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
'''
Stratify parameter will create a equal nube rof samples for train and test
'''

X_train, X_test, y_train, y_test = train_test_split(balanced_data.text, balanced_data.category_number, test_size= 0.2, random_state = 42, stratify = balanced_data.category_number)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```
